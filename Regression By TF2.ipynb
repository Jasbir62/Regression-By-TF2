### Reading Dataset

```python
# importing libraries
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
# using cars data to predict price
pd.options.display.max_columns = 30
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars = pd.read_csv('imports-85.data', header = None, names = cols)
print(cars.shape)
cars
```

```python
# Select only the columns with continuous values from - https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.names
continuous_values_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars = cars[continuous_values_cols]
numeric_cars.info()
```

### Data Cleaning

```python
numeric_cars.head()
```

```python
# convert all data to  numeric data type and perform cleaning
numeric_cars = numeric_cars.replace('?', np.nan)
numeric_cars = numeric_cars.astype('float32')
numeric_cars.isnull().sum()
```

```python
# drop all rows where target column is numm and impute rest values with respective means
numeric_cars = numeric_cars.dropna(subset = ['price'])
numeric_cars.isnull().sum()
```

```python
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
numeric_cars.isnull().sum()
```

```python
# normalize the data to range b/w 0 to 1
price = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min())/(numeric_cars.max() - numeric_cars.min())
numeric_cars['price'] = price
numeric_cars.head()
```

### Predictions with ANN

```python
x_columns = numeric_cars.columns.drop('price')
X = numeric_cars[x_columns]
y = numeric_cars['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
## Using Sequential method
model1 = tf.keras.models.Sequential()
model1.add(tf.keras.layers.Flatten())
model1.add(tf.keras.layers.Dense(64, activation = tf.nn.relu))
model1.add(tf.keras.layers.Dense(48, activation = tf.nn.relu))
model1.add(tf.keras.layers.Dense(1))
model1.compile(optimizer='adam', loss = tf.keras.losses.mse)

model1.fit(X_train, y_train, epochs = 150, batch_size=32, verbose = 0)
error = model1.evaluate(X_test, y_test)
print('MSE: %.3f, RMSE: %.3f' % (error, error **(1/2)))
```

```python
# using Keras functional API
inputs = tf.keras.Input(shape = (len(x_columns,)))
x = tf.keras.layers.Flatten()(inputs)
x = tf.keras.layers.Dense(64, activation = tf.nn.relu, name = 'd1')(x)
x = tf.keras.layers.Dense(64, activation = tf.nn.relu, name = 'd2')(x)
x = tf.keras.layers.Dense(32, activation = tf.nn.relu, name = 'd3')(x)
outputs = tf.keras.layers.Dense(1, name = 'd4')(x)

model2 = tf.keras.Model(inputs, outputs)
model2.compile(optimizer='adam', loss='mse')
model2.fit(X_train, y_train, batch_size=32, epochs=150, verbose=0)
error = model2.evaluate(X_test, y_test)
print('MSE: {:.3f}  RMSE: {:.3f}'.format(error, error**(1/2)))
```
